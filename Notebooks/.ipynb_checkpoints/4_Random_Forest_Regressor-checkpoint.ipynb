```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
import pickle
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
preprocessed_train_path = r'C:\Users\Ben\Desktop\Data Science\Kaggle\Advanced-House-Price-Prediction-Kaggle-Competition\Preprocessed_data\preprocessed_train.csv'
preprocessed_test_path = r'C:\Users\Ben\Desktop\Data Science\Kaggle\Advanced-House-Price-Prediction-Kaggle-Competition\Preprocessed_data\preprocessed_test.csv'
```

```python
df_train = pd.read_csv(preprocessed_train_path)
df_test = pd.read_csv(preprocessed_test_path)
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
df_train.head(10)
```

```python
df_test.head(10)
```

```python
X = df_train.iloc[:, :-1]
y = df_train.iloc[:, -1]
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01, random_state=72)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
rf = RandomForestRegressor()
```

## Model Random Search

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 1200, stop = 4000, num = 15)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt', 'log2']
# Maximum number of levels in tree
max_depth = [3,5,7,9,11,13,15]
# Minimum number of samples required to split a node
min_samples_split = [2,3,4,5,6]
# Minimum number of samples required at each leaf node
min_samples_leaf = [2,3,4,5,6]
# Maximum number of samples to draw from X to train each base estimator.
max_samples = [0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'max_samples': max_samples,
               'criterion': ['mse']}
print(random_grid)
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, scoring='neg_mean_squared_log_error', n_iter = 20, 
                               cv = 5, verbose=10, random_state=72, n_jobs = 1)
```

```python
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_estimator_
```

```python
#Randomized search best score
np.sqrt((-1*rf_random.best_score_))
```

```python
#Train best score
np.sqrt((-1*rf_random.score(X_train, y_train)))
```

```python
#Test best score
np.sqrt((-1*rf_random.score(X_test, y_test)))
```

## Model Grid Search

```python
## Grid Search params
# Number of trees in random forest
n_estimators = [3350, 3400, 3450]
# Number of features to consider at every split
max_features = ['auto']
# Maximum number of levels in tree
max_depth = [13]
# Minimum number of samples required to split a node
min_samples_split = [3]
# Minimum number of samples required at each leaf node
min_samples_leaf = [5]
# Maximum number of samples to draw from X to train each base estimator.
max_samples = [0.65]
```

```python
# Create the complete grid
params = {
            'n_estimators': n_estimators,
            'max_features': max_features,
            'max_depth': max_depth,
            'min_samples_split': min_samples_split,
            'min_samples_leaf': min_samples_leaf,
            'max_samples': max_samples,
            'criterion': ['mse']
         }
print(params)
```

```python
# Set up the grid search with 5-fold cross validation
rf_grid = GridSearchCV(estimator=rf,
            param_grid=params,
            cv=5, 
            scoring = 'neg_mean_squared_log_error',
            verbose = 3,
            n_jobs = -1)
```

```python
rf_grid.fit(X_train, y_train)
```

```python
rf_grid.best_estimator_
```

```python
#Randomized search best score
np.sqrt((-1*rf_grid.best_score_))
```

```python
#Train best score
np.sqrt((-1*rf_grid.score(X_train, y_train)))
```

```python
#Test best score
np.sqrt((-1*rf_grid.score(X_test, y_test)))
```
