```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
import pickle
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
preprocessed_train_path = r'C:\Users\Ben\Desktop\Data Science\Kaggle\Advanced House Price Prediction\Preprocessed_data\preprocessed_train.csv'
preprocessed_test_path = r'C:\Users\Ben\Desktop\Data Science\Kaggle\Advanced House Price Prediction\Preprocessed_data\preprocessed_test.csv'
```

```python
df_train = pd.read_csv(preprocessed_train_path)
df_test = pd.read_csv(preprocessed_test_path)
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
df_train.head(10)
```

```python
df_test.head(10)
```

```python
X = df_train.iloc[:, :-1]
y = df_train.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1
                                                    , random_state=72)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Scaling

```python
scaler = RobustScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
```

```python
df_test = scaler.transform(df_test)
```

## LASSO

```python
alphas_lasso = np.logspace(-30, -1, 30, dtype=float)
```

```python
alphas_lasso
```

```python
lasso_params = {'alpha': alphas_lasso}
```

```python
lasso = Lasso()
```

```python
# Set up the grid search with 5-fold cross validation
lasso_grid = GridSearchCV(estimator=lasso,
            param_grid=lasso_params,
            cv=5, 
            scoring = 'neg_mean_squared_error',
            verbose = 10,
            n_jobs = -1)
```

```python
lasso_grid.fit(X_train, y_train)
```

```python
lasso_grid.best_estimator_
```

```python
np.sqrt((-1*lasso_grid.best_score_))
```

```python
np.sqrt((-1*lasso_grid.score(X_test, y_test)))
```

```python
with open(r'C:\Users\Ben\Desktop\Data Science\Kaggle\Advanced House Price Prediction\Models\lasso', "wb") as f:
    pickle.dump(lasso_grid, f)
```

### Predictions with Model 

```python
with open(r'C:\Users\Ben\Desktop\Data Science\Kaggle\Advanced House Price Prediction\Models\lasso', "rb") as f:
    lasso = pickle.load(f)
```

```python
Id = np.arange(start=1461, stop=2920, step=1)
lasso_submission = pd.DataFrame(data = Id, columns = ['Id'])
```

```python
lasso_prediction = lasso.predict(df_test)
```

```python
lasso_submission['SalePrice'] = np.expm1(lasso_prediction)
```
